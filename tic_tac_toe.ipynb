```python
import random

class tic_tac_toe:

  def __init__(self):
        self.board = []

  def board_creation(self):
    # create the board
    rows , cols = 3,3
    row = []
    for i in range(rows):
      col = []
      for j in range(cols):
        col.append("-")
      self.board.append(col)

  def display_board(self):
    print("\nYour Tic-tac-toe Playground is here --->\n\n")
    for i in range(0,3):
      for j in range(0,3):
        print("({},{})".format(i+1,j+1),self.board[i][j],end = "\t")
      print("\n")

  def generate_1st_player(self):
    return random.randint(0,1)

  def fix_spot(self, r, c, player):
    self.board[r][c] = player

  def win(self,player):
    # checking row win
    n = len(self.board)
    for i in range(n):
      win = True
      for j in range(n):
        if(self.board[i][j]!= player):
          win = False
          break
      if win == True:
        return win
    # Checking column win
    for i in range(n):
      win = True
      for j in range(n):
        if(self.board[j][i]!= player):
          win = False
          break
      if win == True:
        return win
    # checking diagonals
      win = True
      for i in range(n):
        if(self.board[i][i]!= player):
          win = False
          break
      if win == True:
        return win

      win = True
      for i in range(n):
        if(self.board[i][n-i-1]!= player):
          win = False
          break
      if win == True:
       return win

  def draw(self):
    for self.row in self.board:
      for item in self.row:
        if( item == "-" ):
          return False
    return True

  def change_player(self, player):
    if (player == "O"):
      return "X"
    else:
      return "O"

  def start(self):
    # create the play-board
    self.board_creation()

    # display the empty board
    self.display_board()

    # generate 1st random player
    if(self.generate_1st_player() == 0):
      player = "O"
    else:
      player = "X"

    while True:
      print("Turn for player",player)
      # fix move for a player
      r,c = list(map(int,input("\nEnter the row and column of the block you want to put a move in: ").split()))
      self.fix_spot(r-1,c-1,player)

      # checking win
      if(self.win(player) == True):
        self.display_board()
        print()
        print("Player",player,"wins !!!")
        break

      # checking draw
      if(self.draw() == True):
        self.display_board()
        print()
        print("This match is a Draw !!!")
        break

      # changing player turn
      player = self.change_player(player)

      # showing the final view of board
      print()
      self.display_board()

ttt = tic_tac_toe()
ch = input("Do you want to start the game ? (y/n) : ")
if (ch == "y"):
  ttt.start()
else:
  print("Come back next time...Thank you.")
```
